```python
import sqlite3
```

```python
import pandas as pd
```

```python
import hashlib
def hash_symbol(s):
    "map symbol to hex so that quote data are stored in one of 16 sqlite db files"
    return '{:x}'.format(int(hashlib.sha256(s.upper().encode("utf-8")).hexdigest(), 16) % 16)
```

```python
symbol = "googl"
```

```python
db_idx = hash_symbol(symbol)
db_idx
```

```python
db_name = f"quotes_{db_idx}.sqlite"
db_name
```

```python
conn = sqlite3.connect(db_name)
```

```python
c = conn.cursor()
```

```python
# Create table
c.execute('''CREATE TABLE quote_eod
    (TICKER text,
        PER text,
        DATE text,
        TIME text,
        OPEN real,
        HIGH real,
        LOW real,
        CLOSE real,
        VOL INTEGER,
        OPENINT INTEGER
    )
''')
```

```python
conn.commit()
conn.close()
```

```python
table_name = "quote_eod"
```

```python
conn = sqlite3.connect(db_name)
```

```python
csvfile = "googl.us.txt"
```

```python
df = pd.read_csv(csvfile)
```

```python
col_dict = {'<TICKER>': 'TICKER',
     '<PER>': 'PER',
     '<DATE>': 'DATE',
     '<TIME>': 'TIME',
     '<OPEN>': 'OPEN',
     '<HIGH>': 'HIGH',
     '<LOW>': 'LOW',
     '<CLOSE>': 'CLOSE',
     '<VOL>': 'VOL',
     '<OPENINT>': 'OPENINT'}

df.columns = [col_dict.get(x, x) for x in df.columns]
```

```python
df.head()
```

```python
df.to_sql(table_name, conn, if_exists='append', index=False)
```

```python
c = conn.cursor()
```

```python
c.execute(f'SELECT * FROM {table_name} order by TICKER,DATE limit 10;')
```

```python
conn2 = sqlite3.connect("shared.sqlite")
c2 = conn2.cursor()
c2.execute('''CREATE TABLE tickers
    (TICKER text,
        DESC text,
        IND_GROUP text,
        SECTOR text,
        DB_IDX text,
        IN_ACTIVE text,
        EXCHANGE text
    )
''')
```

```python
with open("us-tickers.txt") as f:
    data = [i.split(",") for i in f.read().split("\n")]
```

```python
data
```

```python
col_vals = [(i[0],'', '', '', hash_symbol(i[0]), '', i[1]) for i in data]
```
